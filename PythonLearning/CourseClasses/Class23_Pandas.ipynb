### Iterations in Pandas

```python
df=pd.DataFrame({
    'Name':['Irwin','Tanya','Shubham','Dev'],
    'Age':[20,30,40,50],
    'City':['Delhi','Gurgaon','Noida','Delhi']
})

```

```python
df
```

```python
for x,y in df.iterrows():
    print(x) # first argument takes index
    # second argument takes row data
    print(y.Name)
    print(y.City)
    print(f'{x} index and {y.Name} name and {y.City} city')
```

```python
for row in df.itertuples():
    print(row.Name)
```

```python
df.columns
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())  # unique values in each column
    print(df[col].value_counts())  # count of unique values in each column
    print("-----")
```

```python
for col in df.columns:
    print(col)
    print(df['Name'].unique())  # unique values in each column
    print(df["Name"].value_counts())  # count of unique values in each column
    print("-----")
```
