In Python, a daemon thread is a background thread that runs in the background and automatically exits when the main program ends.

A daemon thread is like a helper or background worker that doesn't block your program from exiting.



ðŸ§  Why Use Daemon Threads?
Useful for tasks like logging, monitoring, or auto-saving, where you donâ€™t need them to finish if the main work is done.

Saves effort in managing when to stop the thread â€” Python handles it automatically.

```python
import threading
import time

def background_task():
    while True:
        print("Daemon thread is running...")
        time.sleep(1)

# Create a daemon thread
t = threading.Thread(target=background_task)
t.daemon = True     # Mark as daemon
t.start()

print("Main program is done!")
```

âš ï¸ Important:
Non-daemon threads keep the program alive.

Daemon threads die silently when the main program finishes.

### Daemon Threads vs Control (Non-Daemon) Threads 

```python
##Code Comparison
import threading
import time

def worker():
    for i in range(5):
        print(f"Worker running: {i}")
        time.sleep(1)

# ðŸ”¸ Daemon Thread
daemon_thread = threading.Thread(target=worker)
daemon_thread.daemon = True
daemon_thread.start()

# ðŸ”¹ Control (Non-Daemon) Thread
control_thread = threading.Thread(target=worker)
control_thread.start()

print("Main thread done!")
```

ðŸ§ª Behavior
ðŸ”¸ If you run the daemon thread only:
Output stops immediately after "Main thread done!" â€” even if the worker is still running.

The daemon thread does not block program exit.

ðŸ”¹ If you run the control thread:
Python waits for it to complete all 5 iterations before exiting the program.
